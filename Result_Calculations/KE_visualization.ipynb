```python
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import pandas as pd
ocean_data=xr.open_dataset("/Users/loganknudsen/Documents/UMD_Research_Local_File_Saves/BBL_w_O_updated_diagnostics_TKE_terms.nc") #i=103
```

```python
fig, axs = plt.subplots()

frms = len(ocean_data.KE.time)
def animate(t):
    plt.clf()
    ocean_data.KE[t].plot(add_colorbar=True,norm=matplotlib.colors.LogNorm(vmin=10**(-10),vmax=10**(-3)),cmap='bwr') # vmin=-0.001,vmax=0.001
    # plt.contour(b[t,:,0,:],cmap="Greys")
    # b[t*10,:,0,:].plot.contour(cmap="ocean",levels= 10)

# matplotlib.animation.FuncAnimation(fig, animate, frames=frms)
anim = matplotlib.animation.FuncAnimation(fig, animate, frames=frms)
# # writervideo = matplotlib.animation.FFMpegWriter(fps=60) 
anim.save('KE_plot.gif',fps=480) 
plt.close()
```

```python
t = 3906
tm = ocean_data.time[t]/pd.Timedelta("1s")*(1e-8+1e-4*1e-5)**(0.5)/(2*np.pi)
tm.data
```

```python
t = 1971 # 1971 3906
tm = round(ocean_data.time[t].data/pd.Timedelta("1s")*(1e-8+1e-4*1e-5)**(0.5)/(2*np.pi),2)
fig, axs = plt.subplots()
ocean_data.KE[t].plot(add_colorbar=True,norm=matplotlib.colors.LogNorm(vmin=10**(-10),vmax=10**(-3)),cmap='bwr',cbar_kwargs={'label': "Pertubation Kinetic Energy ($k[m^2/s^2$])"}) # vmin=-0.001,vmax=0.001
# plt.contour(b[t,:,0,:],cmap="Greys")
# b[t,:,0,:].plot.contour(cmap="black",levels= 10)
plt.title("Pertubation kinetic energy $k$ at $t\dfrac{f^*}{2\pi}=$"+str(tm))
plt.xlabel("Distance along-slope($y[m]$)")
plt.ylabel("Height above slope($z[m]$)")
plt.show()
tm
```
